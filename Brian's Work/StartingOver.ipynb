```python
from bs4 import BeautifulSoup
import pandas as pd
import re

# Load HTML
with open("Diablo_IV_Patch_Notes.html", "r", encoding="utf-8") as f:
    soup = BeautifulSoup(f, "html.parser")

data = []

# Each patch is inside a div.panel
panels = soup.find_all("div", class_="panel")

for panel in panels:

    # --- Extract Title Block ---
    title_div = panel.find("div", class_="panel-title")
    if not title_div:
        continue

    title_text = title_div.get_text(strip=True)

    # --- Extract Version ---
    version_match = re.search(r"\d+\.\d+\.\d+", title_text)
    version = version_match.group(0) if version_match else "UNKNOWN"

    # --- Extract Build ---
    build_match = re.search(r"Build\s+#?(\d+)", title_text)
    build = build_match.group(1) if build_match else "UNKNOWN"

    # --- Extract Date (embedded in title text) ---
    date_match = re.search(
        r"(January|February|March|April|May|June|July|August|September|October|November|December)\s+\d{1,2},\s+\d{4}",
        title_text
    )
    date = date_match.group(0) if date_match else "UNKNOWN"

    # --- Extract Patch Body ---
    body = panel.find("div", class_="panel-body")
    if not body:
        continue

    # Extract true bullet points
    bullets = body.find_all("li")

    for li in bullets:
        text = li.get_text(separator=" ", strip=True)

        if text:
            data.append({
                "version": version,
                "build": build,
                "date": date,
                "bullet_text": text
            })

# Create DataFrame
df_bullets = pd.DataFrame(data)

print("Total bullets extracted:", len(df_bullets))

# Verify structure
display(
    df_bullets[["version", "build", "date"]]
    .drop_duplicates()
    .sort_values("version")
)
```

```python
import re

# -----------------------------
# 1. BUGFIX DETECTION
# -----------------------------
def detect_bugfix(text):
    text_lower = text.lower()
    bugfix_signals = [
        "fixed an issue",
        "resolved an issue",
        "corrected an issue",
        "addressed an issue",
        "fixed an amount"
    ]
    return any(signal in text_lower for signal in bugfix_signals)


# -----------------------------
# 2. UI IMPROVEMENT DETECTION
# -----------------------------
def detect_ui(text):
    text_lower = text.lower()
    ui_keywords = [
        "ui",
        "tooltip",
        "screen reader",
        "display",
        "visual indicator",
        "menu",
        "map",
        "icon",
        "clarity",
        "text updated",
        "description updated"
    ]
    return any(word in text_lower for word in ui_keywords)


# -----------------------------
# 3. QUALITY OF LIFE DETECTION
# -----------------------------
def detect_qol(text):
    text_lower = text.lower()

    qol_keywords = [
        "now spawns closer",
        "removed the delay",
        "no longer required",
        "unlock earlier",
        "unlock at level",
        "respawn closer",
        "added audio cue",
        "improved accessibility",
        "no longer need",
        "portal now",
        "will never respawn"
    ]

    return any(phrase in text_lower for phrase in qol_keywords)


# -----------------------------
# 4. NUMERIC DIRECTION DETECTION
# -----------------------------
def detect_numeric_direction(text):
    text_lower = text.lower()

    # Detect "from X to Y"
    pattern = r"from\s+(\d+\.?\d*)%?\s+to\s+(\d+\.?\d*)%?"
    matches = re.findall(pattern, text_lower)

    for before, after in matches:
        before = float(before)
        after = float(after)

        if after > before:
            return "Buff"
        elif after < before:
            return "Nerf"

    return None


# -----------------------------
# 5. DIRECTIONAL KEYWORD LOGIC
# -----------------------------
def detect_directional_keywords(text):
    text_lower = text.lower()

    # COST LOGIC
    if "cost reduced" in text_lower:
        return "Buff"
    if "cost increased" in text_lower:
        return "Nerf"

    # DROP RATE LOGIC
    if "rarer" in text_lower:
        return "Nerf"
    if "drop rate increased" in text_lower or "more items now drop" in text_lower:
        return "Buff"

    # MONSTER TUNING
    if "monster" in text_lower or "boss" in text_lower:
        if "increased" in text_lower or "tuned up" in text_lower:
            return "Nerf"
        if "reduced" in text_lower or "tuned down" in text_lower or "lowered" in text_lower:
            return "Buff"

    # GENERAL BUFF/NERF WORDS
    if "increased" in text_lower:
        return "Buff"
    if "reduced" in text_lower:
        return "Nerf"
    if "lowered" in text_lower:
        return "Buff"
    if "removed" in text_lower:
        return "Nerf"

    return None


# -----------------------------
# MAIN CLASSIFIER (HIERARCHICAL)
# -----------------------------
def classify_bullet(text):

    # 1️⃣ Bugfix
    if detect_bugfix(text):
        return "Bugfix"

    # 2️⃣ UI
    if detect_ui(text):
        return "UI_Improvement"

    # 3️⃣ Quality of Life
    if detect_qol(text):
        return "Quality_of_Life"

    # 4️⃣ Numeric Direction
    numeric_result = detect_numeric_direction(text)
    if numeric_result:
        return numeric_result

    # 5️⃣ Keyword Direction
    keyword_result = detect_directional_keywords(text)
    if keyword_result:
        return keyword_result

    # 6️⃣ Fallback
    return "Neutral"
```

```python
import re

# -----------------------------
# 1. BUGFIX
# -----------------------------
def detect_bugfix(text):
    text_lower = text.lower()
    return text_lower.startswith("fixed an issue") or \
           "resolved an issue" in text_lower or \
           "corrected an issue" in text_lower


# -----------------------------
# 2. UI IMPROVEMENT
# -----------------------------
def detect_ui(text):
    text_lower = text.lower()
    ui_keywords = [
        "tooltip", "ui", "screen reader", "visual",
        "display", "menu", "map", "icon", "clarity"
    ]
    return any(word in text_lower for word in ui_keywords)


# -----------------------------
# 3. QUALITY OF LIFE
# -----------------------------
def detect_qol(text):
    text_lower = text.lower()
    qol_keywords = [
        "spawns closer", "removed the delay",
        "no longer required", "unlock earlier",
        "respawn closer", "improved accessibility"
    ]
    return any(word in text_lower for word in qol_keywords)


# -----------------------------
# 4. NUMERIC COMPARISON (Robust)
# -----------------------------
def detect_numeric_direction(text):
    text_lower = text.lower()

    # Detect "from X to Y" including bracket multipliers
    pattern = r"from\s+(\d+\.?\d*)[^\d]*\s+to\s+(\d+\.?\d*)"
    matches = re.findall(pattern, text_lower)

    for before, after in matches:
        before = float(before)
        after = float(after)

        if after > before:
            return "Buff"
        elif after < before:
            return "Nerf"

    # Detect "up from X"
    up_match = re.search(r"up from\s+(\d+\.?\d*)", text_lower)
    if up_match:
        return "Buff"

    # Detect "down from X"
    down_match = re.search(r"down from\s+(\d+\.?\d*)", text_lower)
    if down_match:
        return "Nerf"

    return None


# -----------------------------
# 5. STAT LINE DETECTION
# -----------------------------
def detect_stat_line(text):
    stripped = text.strip()
    if stripped.startswith("+"):
        return "Buff"
    return None


# -----------------------------
# 6. SEMANTIC DIRECTION
# -----------------------------
def detect_semantic_direction(text):
    text_lower = text.lower()

    # Cap removal
    if "no longer has a cap" in text_lower:
        return "Buff"

    # Cap added (only if explicit comparison absent)
    if "now has a cap" in text_lower:
        return "Nerf"

    # Positive framing
    if "more balanced" in text_lower or "better balanced" in text_lower:
        return "Buff"

    # Cost changes
    if "cost reduced" in text_lower:
        return "Buff"
    if "cost increased" in text_lower:
        return "Nerf"

    # Drop rate changes
    if "rarer" in text_lower:
        return "Nerf"
    if "more items now drop" in text_lower:
        return "Buff"

    # Monster tuning
    if "monster" in text_lower or "boss" in text_lower:
        if "increased" in text_lower or "tuned up" in text_lower:
            return "Nerf"
        if "reduced" in text_lower or "tuned down" in text_lower or "lowered" in text_lower:
            return "Buff"

    # General keywords (conservative)
    if "increased" in text_lower:
        return "Buff"
    if "reduced" in text_lower:
        return "Nerf"
    if "lowered" in text_lower:
        return "Buff"

    return None


# -----------------------------
# MAIN CLASSIFIER (HIERARCHICAL)
# -----------------------------
def classify_bullet(text):

    # 1️⃣ Bugfix
    if detect_bugfix(text):
        return "Bugfix"

    # 2️⃣ UI
    if detect_ui(text):
        return "UI_Improvement"

    # 3️⃣ QoL
    if detect_qol(text):
        return "Quality_of_Life"

    # 4️⃣ Stat lines (+ prefix)
    stat_result = detect_stat_line(text)
    if stat_result:
        return stat_result

    # 5️⃣ Numeric comparison
    numeric_result = detect_numeric_direction(text)
    if numeric_result:
        return numeric_result

    # 6️⃣ Semantic direction
    semantic_result = detect_semantic_direction(text)
    if semantic_result:
        return semantic_result

    # 7️⃣ Fallback
    return "Neutral"
```

```python
# -----------------------------
# 1. BUGFIX
# -----------------------------
def detect_bugfix(text):
    t = text.lower()
    return t.startswith("fixed an issue") or \
           "resolved an issue" in t or \
           "corrected an issue" in t


# -----------------------------
# 2. UI IMPROVEMENT
# -----------------------------
def detect_ui(text):
    t = text.lower()
    ui_keywords = [
        "tooltip", "ui", "screen reader",
        "display", "visual", "menu",
        "map", "icon", "clarity"
    ]
    return any(word in t for word in ui_keywords)


# -----------------------------
# 3. QUALITY OF LIFE
# -----------------------------
def detect_qol(text):
    t = text.lower()
    qol_keywords = [
        "spawns closer",
        "removed the delay",
        "unlock earlier",
        "respawn closer",
        "added audio cue",
        "improved accessibility"
    ]
    return any(word in t for word in qol_keywords)


# -----------------------------
# 4. NUMERIC DIRECTION (Robust)
# -----------------------------
def detect_numeric_direction(text):
    t = text.lower()

    # from X to Y (handles multipliers like 4[x])
    pattern = r"from\s+(\d+\.?\d*)[^\d]*\s+to\s+(\d+\.?\d*)"
    matches = re.findall(pattern, t)

    for before, after in matches:
        before = float(before)
        after = float(after)

        if after > before:
            return "Buff"
        elif after < before:
            return "Nerf"

    # up from
    if re.search(r"up from\s+\d+\.?\d*", t):
        return "Buff"

    # down from
    if re.search(r"down from\s+\d+\.?\d*", t):
        return "Nerf"

    return None


# -----------------------------
# 5. STAT LINE (+ prefix)
# -----------------------------
def detect_stat_line(text):
    stripped = text.strip()
    if stripped.startswith("+"):
        return "Buff"
    return None


# -----------------------------
# 6. REMOVAL LOGIC (Object-sensitive)
# -----------------------------
def detect_removal_logic(text):
    t = text.lower()

    if "removed" not in t and "no longer" not in t:
        return None

    negative_objects = ["requirement", "delay", "hazard"]
    positive_objects = [
        "damage", "bonus", "resource",
        "healing", "resistance",
        "cooldown reduction", "armor"
    ]

    # Removed negative thing → Buff
    if any(obj in t for obj in negative_objects):
        return "Buff"

    # Removed positive thing → Nerf
    if any(obj in t for obj in positive_objects):
        return "Nerf"

    # Ambiguous removal → Neutral
    return None


# -----------------------------
# 7. CAP LOGIC
# -----------------------------
def detect_cap_logic(text):
    t = text.lower()

    if "no longer has a cap" in t:
        return "Buff"

    if "now has a cap" in t:
        return "Nerf"

    return None


# -----------------------------
# 8. MONSTER / ENEMY INVERSION
# -----------------------------
def detect_monster_inversion(text):
    t = text.lower()

    if any(word in t for word in ["monster", "boss", "enemy", "hazard"]):
        if "increased" in t or "tuned up" in t:
            return "Nerf"
        if "reduced" in t or "lowered" in t or "tuned down" in t:
            return "Buff"

    return None


# -----------------------------
# 9. POSITIVE FRAMING
# -----------------------------
def detect_positive_framing(text):
    t = text.lower()

    if "more balanced" in t or "better balanced" in t:
        return "Buff"

    return None


# -----------------------------
# 10. GENERAL DIRECTION KEYWORDS
# -----------------------------
def detect_general_keywords(text):
    t = text.lower()

    if "increased" in t:
        return "Buff"
    if "reduced" in t:
        return "Nerf"
    if "lowered" in t:
        return "Buff"

    return None


# -----------------------------
# MAIN CLASSIFIER (HIERARCHICAL)
# -----------------------------
def classify_bullet(text):

    # 1️⃣ Bugfix
    if detect_bugfix(text):
        return "Bugfix"

    # 2️⃣ UI
    if detect_ui(text):
        return "UI_Improvement"

    # 3️⃣ QoL
    if detect_qol(text):
        return "Quality_of_Life"

    # 4️⃣ + stat lines
    stat_result = detect_stat_line(text)
    if stat_result:
        return stat_result

    # 5️⃣ Numeric comparison
    numeric_result = detect_numeric_direction(text)
    if numeric_result:
        return numeric_result

    # 6️⃣ Removal logic
    removal_result = detect_removal_logic(text)
    if removal_result:
        return removal_result

    # 7️⃣ Cap logic
    cap_result = detect_cap_logic(text)
    if cap_result:
        return cap_result

    # 8️⃣ Monster inversion
    monster_result = detect_monster_inversion(text)
    if monster_result:
        return monster_result

    # 9️⃣ Positive framing
    framing_result = detect_positive_framing(text)
    if framing_result:
        return framing_result

    # 🔟 General keywords (last resort)
    keyword_result = detect_general_keywords(text)
    if keyword_result:
        return keyword_result

    return "Neutral"
```

```python
df_bullets["impact_category"] = df_bullets["bullet_text"].apply(classify_bullet)

df_bullets["impact_category"].value_counts()
```

```python
from transformers import pipeline
import torch
import pandas as pd
from tqdm import tqdm

# Use GPU if available
device = 0 if torch.cuda.is_available() else -1

classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli",
    device=device
)

candidate_labels = [
    "Buff",
    "Nerf",
    "Bugfix",
    "UI Improvement",
    "Quality of Life",
    "Neutral"
]
```

```python
def classify_zero_shot(text):
    result = classifier(
        text,
        candidate_labels,
        multi_label=False
    )
    return result["labels"][0]

tqdm.pandas()

df_bullets["impact_category"] = df_bullets["bullet_text"].progress_apply(classify_zero_shot)

df_bullets["impact_category"].value_counts()
```
